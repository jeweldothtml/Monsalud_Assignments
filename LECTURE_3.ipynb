Graphs

A **Graph** is a fundamental mathematical structure used to represent pairwise relations between nodes which are vertices and their links being called edges

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
graph = nx.Graph()
graph.add_edge(1, 2)
graph.add_node(3)
graph.add_edge(3, 3)
nx.draw_spring(graph, with_labels = True)
plt.title("G")
plt.show()
```

A **Directed Graph** have edges which have specific directs from where they emerge and link to. They can be seen as arrows, an example in the graph below would be, C to A.

A **Simple Graph** is an *undirected, unweighted, no multiple edges, and no loops* graph.

A **Complete Graph** is an undirected graph in which every pair of distinct vertices is connected by a unique edge. In other words every vertex and edge is adjacent to all other vertex and edges.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
comgra = nx.complete_graph(5)
nx.draw_spring(comgra, with_labels = True)
plt.title("Complete Graph")
plt.show()
```

A **Bipartite Graph** is a graph in which its vertices are divided into two disjoint and independent sets. Every Vertex in either set is not adjacent to the vertex in the same set, however every edge connects vertices from one set to the other

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
bigra = nx.Graph()
setA = [1, 2, 3, 4]
setB = [7, 8, 9]
bigra.add_nodes_from(setA, bipartite=0)
bigra.add_nodes_from(setB, bipartite=1)
for nodeA in setA:
    for nodeB in setB:
        bigra.add_edge(nodeA, nodeB)
pos = {node: (0, i) for i, node in enumerate(setA)}
pos.update({node: (1, i) for i, node in enumerate(setB)})
nx.draw(bigra, pos, node_color='skyblue', node_size=500, with_labels=True, edge_color='gray', width=1.5)
plt.title("Bipartite Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"})
exgra = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("Main Graph")
plt.show()
print("Degree of each vertex: ")
print(dict(G.degree))
```

**Handshake Theorem** <br>
In any undirected graph, the theorem states that the sum of the degrees of all vertices is always equal to twice the number of edges in the Graph.

```python

import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"})
exgra = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("Main Graph")
plt.show()
print("Twice number of edges: ")
print(nx.number_of_edges(G) * 2)
degrees = G.degree()
result = sum(deg for node, deg in degrees)
print("Sum of degrees of all vertices: ")
print(result)
```

**Travelling in a Graph**

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"})
exgra = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("Main Graph")
plt.show()
```

A **Walk** from vertex 'A' to 1 follows the edge to X, then to 4, 3, 2, arriving to vertex 1.  A → X → 4 → 3 → 2 → 1. <br>
A **Trail** from vertex 'A' to 3 follows the edge to X, then 4, arriving to 3.  A → X → 4 → 3. Note that a trail contains no repeated edge. <br>
A **Path** from vertex 'A' to 3 follows the edge to X, then 4, arriving to 3.  A → X → 4 → 3. Note that a path contains no repeated vertex. <br>
A **Closed Walk** from vertex 'A' follows the edge to X, then 4, 3, then arriving to 'A'.  A → X → 4 → 3 → A. Note that a closed walk starts and ends with the same vertex. <br>
A **Circuit** from vertex 'B' follows the edge to 'A', then to 'X' then back to 'B'.  B → A → X → B. Note that a circuit is a *closed trail*. <br>
A **Simple Circuit** from vertex 'X' follows the edge to 'A', then to 'B' back to 'X'. X → A → B → X. Note that a simple circuit is a *closed path*, that starts and ends with the same vertex, without any repeating edges and vertices except from the source and end vertex. <br>

**Euler's Circuit**

2 → 4 → 3 → 2 → 0 → 1 →2

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
g = nx.Graph()
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(2, 4)
g.add_edge(3, 4)
nx.draw_spring(g, with_labels = True)
plt.title("Euler's Circuit")
plt.show()
```

**Euler's Trail**

0 → 4 → 5 → 1 → 6 → 5 → 2 → 3 → 7 → 2 → 6 → 3 → 5 → 0

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
G = nx.Graph()
edges = [(0, 4), (4, 5), (5, 1), (1, 6), (6, 5), (5, 2), (2, 3), (3, 7), (7, 2), (2, 6), (6, 3), (3, 5), (5, 0)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10)
plt.title("Eulerian Trail")
plt.show()
```

**Hammiltonian Circuit**

0 → 1 → 2 → 3 → 0

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
G = nx.Graph()
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (0, 2), (1, 3)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Hamiltonian Circuit")
plt.show()
```

**Travelling Salesman Problem**

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
G.add_edge("A", "B", weight=1.0)
G.add_edge("B", "C", weight=2.0)
G.add_edge("A", "C", weight=3.0)
G.add_edge("A", "D", weight=1.5)
G.add_edge("D", "E", weight=2.5)
G.add_edge("E", "F", weight=0.5)
G.add_edge("B", "F", weight=1.2)
G.add_edge("C", "F", weight=1.7)
G.add_edge("C", "E", weight=0.8)
G.add_edge("F", "A", weight=0.9)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
